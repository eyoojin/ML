# 선형 회귀

```python
import pandas as pd

df =pd.read_csv('data/Fish.csv')
```

```python
perch_df = df.loc[df['Species'] == 'Perch'] 
```

```python
perch_length = perch_df[['Length2']] # 데이터를 2차원으로
perch_weight = perch_df[['Weight']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(train_input, train_target)
```

```python
lr.score(test_input, test_target)
```

```python
lr.predict([[100]])
```

```python
lr.coef_ # coefficient # ax + b에서 a 
# 기울기 -> 계수(coefficient) 또는 **가중치(weight)**: 영향을 얼마나 줄까에 대한 정보
```

```python
lr.intercept_ # ax + b에서 b
```

```python
import matplotlib.pyplot as plt

plt.scatter(train_input, train_target)
plt.plot([15, 50], [15 * 37 - 657, 50 * 37 - 657])
```

```python
print(lr.score(train_input, train_target))
print(lr.score(test_input, test_target))
```

```python
lr.predict([[5]]) # 선형회귀 모델의 부족한 점
```
