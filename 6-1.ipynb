# 군집 알고리즘

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fruits = np.load('data/fruits_300.npy')
```

```python
fruits.shape # 3차원 데이터, 사진 300장/ 100 X 100 사진
```

```python
fruits[0, 0] # 첫번째 사진의 맨 윗줄
```

```python
plt.imshow(fruits[0], cmap='gray')
plt.show()
```

```python
fig, axs = plt.subplots(1, 2)
axs[0].imshow(fruits[100])
axs[1].imshow(fruits[200])
plt.show()
```

```python
apple = fruits[0:100].reshape(-1, 100*100) # 3차원을 2차원으로 만드는 작업
pineapple = fruits[100:200].reshape(-1, 100*100)
banana = fruits[200:300].reshape(-1, 100*100)
```

```python
apple.shape
```

```python
apple.mean(axis=1)
```

```python
plt.hist(np.mean(apple, axis=1))
plt.hist(np.mean(pineapple, axis=1))
plt.hist(np.mean(banana, axis=1))
plt.show()
```

```python
apple_mean = np.mean(apple, axis=0).reshape(100, 100)
pineapple_mean = np.mean(pineapple, axis=0).reshape(100, 100)
banana_mean = np.mean(banana, axis=0).reshape(100, 100)
```

```python
abs_diff = np.abs(fruits - apple_mean)
abs_mean = np.mean(abs_diff, axis=(1, 2))
```

```python
apple_index = np.argsort(abs_mean)[0:100]

fig, axs = plt.subplots(10, 10, figsize=(10, 10))
for i in range(10):
    for j in range(10):
        axs[i, j].imshow(fruits[apple_index[i*10 + j]], cmap='gray')
        axs[i, j].axis('off')

plt.show()
```
